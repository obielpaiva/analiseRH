```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r'BaseFuncionarios.xlsx')
df.head()
```

O .shape retorna uma tupla com as dimensões do DataFrame:
| Código        | Retorna o quê?        | Significado                       |
| ------------- | --------------------- | --------------------------------- |
| `df.shape`    | `(linhas, colunas)`   | Tamanho da tabela                 |
| `df.shape[0]` | Número de **linhas**  | Quantidade de registros           |
| `df.shape[1]` | Número de **colunas** | Quantidade de atributos/variáveis |

```python
#transformando as colunas 'data de contratação' e 'data de demissão' para o formato de data
df['Data de Contratacao'] = pd.to_datetime(df['Data de Contratacao'])
df['Data de Demissao'] = pd.to_datetime(df['Data de Demissao'], errors='coerce') #caso haja data sem número ou com NaN, ele não exibe erro
```

```python
#cria a coluna status, #método "isna" retorna True ou False, Caso esteja "False" o funcionário está ativo
df['Status'] = df['Data de Demissao'].isna().map({True:'Ativo', False: 'Desligado'})# Se estive True, ele muda para ativo caso esteja False ele muda para Desligado
df['Status'].value_counts() #mostra quantos funcionários estão ativos ou desligados
```

```python
total_ativos = df[df['Status']== 'Ativo'].shape[0]
total_desligados = df[df['Status']== 'Desligado'].shape[0]
print(total_ativos)
print(total_desligados)
```

```python
total_contratacoes = df['Data de Contratacao'].notna().sum()#'notna()' retorna se o valor não for nulo ele faz algo e o .sum() soma
print(total_contratacoes)
df.head()
```

```python
df['Folha salarial'] = df['Salario'] + df['VR'] + df['VT']
total_folha = df[df ['Status']=='Ativo']['Folha salarial'].sum()
df.head()
#outra forma
folha_salarial = (df['Salario'] + df['VR'] + df['VT']) [df['Status']=='Ativo']['Folha salarial'].sum()
```
